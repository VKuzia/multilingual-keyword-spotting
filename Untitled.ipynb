```python
language = 'ru'
# PATH_TO_WAV = f"dataset/multilingual_spoken_words/wav/{language}/clips"
PATH_TO_WAV = f'./'

labels =  ['прыг']
print(labels)
```

```python
waveforms = []
for label in labels:
    items = os.listdir(f'{PATH_TO_WAV}/{label}')
    for item in items:
        if item.startswith('x'):
            wav, sample_rate = torchaudio.load(f'{PATH_TO_WAV}/{label}/{item}')
            print(wav, sample_rate)
            print(wav.shape)
            waveforms.append(wav)
```

```python
resampler = torchaudio.transforms.Resample(44100, 16000)
```

```python
fig, axes = plt.subplots(len(waveforms), 1, figsize=(5, 40))
monos = []
for ax, wav in zip(axes, waveforms):
#     ax.plot(wav)
    print(wav.shape)
    resampled = resampler(wav)
    print(resampled.shape)
    print(resampled[0])
    mono = (resampled[0]) / 2
    ax.hist([mono])
    monos.append(mono)
#     ax.hist(wav[0, :])
#     resampled = resampler(wav)
#     ax.hist(resampled[0, :])
#     print(resampled.shape)
fig.show()
```

```python
print(monos, len(monos))
```

```python
def tensor_to_img(spectrogram):
    plt.figure(figsize=(14,5)) # arbitrary, looks good on my screen.
    plt.imshow(spectrogram[0])
    plt.show()
    display(spectrogram.shape)
```

```python
transform = torchaudio.transforms.MelSpectrogram(n_mels=49, hop_length=401, power=1)
```

```python
for mono in monos:
    spec = transform(mono)
#     print(spec)
    tensor_to_img(spec.unsqueeze(0))
```

```python
for i in range(len(monos)):
    torchaudio.save(f'temp/y_{i}.wav', monos[i].unsqueeze(0), 16000)
    print(f'y_{i}.wav saved')
```
